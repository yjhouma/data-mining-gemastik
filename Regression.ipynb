```python
import pandas as pd
```

```python
data = pd.read_csv("forestfires.csv")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data["area_log"] = np.log(data["area"] + 1)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(211)
plt.hist(data["area"])

plt.subplot(212)
plt.hist(data["area_log"])

plt.show()
```

```python
hehe = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
def month_to_season(x):
    if x in ['mar','apr','may']:
        return 'spring'
    elif x in ['jun', 'jul', 'aug']:
        return 'summer'
    elif x in ['sep', 'oct', 'nov']:
        return 'autumn'
    elif x in ['dec', 'jan', 'feb']:
        return 'winter'
    else:
        raise ValueError("Mantap!!!")
```

```python
data["season"] = data["month"].apply(lambda x: month_to_season(x))
data
```

```python
data.drop("month",1,inplace=True)
```

```python
def make_fire(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
data["fire_exist"] = data["area"].apply(lambda x: make_fire(x))
data.head()
```

```python
sns.countplot(data["season"],hue=data["fire_exist"])
plt.show()
```

```python
data.drop("day", 1,inplace=True)
```

```python
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun','jul','aug','sep','oct','nov','dec']
days = ['sun','mon','tue','wed','thu','fri','sat']
```

```python
def change_months(x):
    return months.index(x)

def change_days(x):
    return days.index(x)
```

```python
data["month"] = data["month"].apply(lambda x: change_months(x))
data["day"] = data["day"].apply(lambda x: change_days(x))
data.head()
```

```python
#PREPROCESSES DULU BRO
def one_hot_encode(dataframe, nama_kolom):
    onehot = pd.get_dummies(dataframe[nama_kolom],prefix=nama_kolom)
    ret = dataframe.drop(nama_kolom,1)
    return ret.join(onehot)
    
```

```python
for i in ["season"]:
    data = one_hot_encode(data,i)
data
```

```python
X = data.drop(["area","area_log","X",'Y'],1)
y = data["area_log"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
from sklearn.cross_validation import StratifiedKFold
from sklearn.metrics import mean_squared_error
def crossvalModel(model,X,y):
    err =[]
    kf = StratifiedKFold(y=y,n_folds=10,shuffle=True,random_state=10)
    for train_index, test_index in kf:
        train_X = X[train_index, :]
        train_Y = y.iloc[train_index].values
        test_X = X[test_index, :]
        test_Y = y.iloc[test_index].values
        
        #train model
        model.fit(train_X, train_Y)
        pred = model.predict(test_X)
        err.append(mean_squared_error(test_Y, pred)**.5)
    return err
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model1 = RandomForestRegressor()
model2 = SVR(kernel='linear')
model3 = SVR(kernel='rbf')
# Create ANN with one hidden layer
model4 = Sequential()
model4.add(Dense(units=16, input_dim=X.shape[1],activation='relu'))
model4.add(Dropout(.15))
model4.add(Dense(units=1, activation='linear'))
model4.compile(optimizer="adam", loss="mean_squared_error")
model5 = KNeighborsRegressor()
```

```python
all_model = [model1,model2,model3,model4,model5]
allerr = {}
index=0
for i in all_model:
    allerr[index] = crossvalModel(i,X,y)
    index+=1
```

```python
allerr
```

```python
allerr = pd.DataFrame(allerr)
allerr.columns = ["Random_Forrest","SVM_linear","SVM_rbf","ANN"
                  ,"KNN"]
allerr.index = range(1,11)
```

```python
allerr.plot(figsize=(10,10))
#axes = plt.gca()
#axes.set_xlim([3,10])
#axes.set_ylim([0,1.5])
plt.xlabel("Number of Fold")
plt.ylabel("RMSLE")
plt.show()
```
