```python
import pandas as pd
```

```python
data = pd.read_csv("forestfires.csv")
data.head()
```

```python
def make_fire(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
data["fire_exist"] = data["area"].apply(lambda x: make_fire(x))
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data["fire_exist"])
plt.show()
```

```python
data.drop("area",axis=1,inplace=True)
```

```python
data.head()
```

```python
def one_hot_encode(dataframe, nama_kolom):
    onehot = pd.get_dummies(dataframe[nama_kolom],prefix=nama_kolom)
    ret = dataframe.drop(nama_kolom,1)
    return ret.join(onehot)
```

```python
for i in ["month","day"]:
    data = one_hot_encode(data,i)
data.head()
```

```python
tSNE_vect = pd.DataFrame(tSNE_vect,
                        columns= ["t-SNE_X", "t-SNE_y"])
```

```python
tSNE_vect = tSNE_vect.join(data["fire_exist"])
```

```python
sns.lmplot('t-SNE_X',
            't-SNE_y',
          data=tSNE_vect,
          fit_reg=False,
          hue='fire_exist')
plt.show()
```

```python
from sklearn.cross_validation import StratifiedKFold
from sklearn.metrics import recall_score
def crossvalModel(model,X,y,ann=False):
    err =[]
    kf = StratifiedKFold(y=y,n_folds=10)
    for train_index, test_index in kf:
        train_X = X[train_index, :]
        train_Y = y[train_index]
        test_X = X[test_index, :]
        test_Y = y[test_index]
        
        #train model
        model.fit(train_X, train_Y)
        if not ann:
            pred = model.predict(test_X)
        else:
            pred = model.predict_classes(test_X)
        err.append(recall_score(test_Y, pred))
        print(err)
    return err
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop("fire_exist",1).values
y = data["fire_exist"].values
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model1 = RandomForestClassifier()
model2 = GradientBoostingClassifier()
model3 = SVC()
model4 = LogisticRegression()
model5 = Sequential()
model5.add(Dense(units=16, input_dim=X.shape[1],activation='relu'))
model5.add(Dropout(.15))
model5.add(Dense(units=1, activation='sigmoid'))
model5.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
all_model = [model1,model2,model3,model4,model5]
allerr = {}
index=0
for i in all_model:
    if index <4:
        allerr[index] = crossvalModel(i,X,y)
    else:
        allerr[index] = crossvalModel(i,X,y,ann=True)

    index+=1
    
allerr = pd.DataFrame(allerr)
```

```python
allerr.columns = ["Random_Forest","GradientBoosting","SVC","LogisticReg","ANN"]
```
